## Data Preprocessing

### OverSampling

```python
oversample = RandomOverSampler()
y = data["DEATH_EVENT"].factorize()[0]
data.drop(columns=["DEATH_EVENT"],inplace=True)
X,Y = oversample.fit_resample(data,y)
print("The number of the survival are {}. The number of the dead are {}.".format(np.count_nonzero(Y),len(Y)-np.count_nonzero(Y)))
```

```python
# Proportion of categories after oversamepling 
new_Y = pd.Series(Y)
new_df = pd.concat([X,new_Y],axis=1)
new_df.rename(columns= {0:"DEATH_EVENT"},inplace=True)
new_df["DEATH_EVENT"] = new_df["DEATH_EVENT"].replace({0: "Survival", 1: "Death"})
new_df["sex"] = new_df["sex"].replace({0: "Female", 1: "Male"})
new_df["anaemia"] = new_df["anaemia"].replace({0: "False", 1: "True"})
new_df["high_blood_pressure"] = new_df["high_blood_pressure"].replace({0: "False", 1: "True"})
new_df["diabetes"] = new_df["diabetes"].replace({0: "False", 1: "True"})
new_df["smoking"] = new_df["smoking"].replace({0: "False", 1: "True"})
alt.Chart(new_df).mark_rect().encode(
    y=alt.Y('DEATH_EVENT:N',title=None),
    x=alt.X(alt.repeat("column"), type='nominal'),
    color='count(DEATH_EVENT):Q'
).properties(
    width=100,
    height=100
).repeat(column=qualitative_columns)
```

```python
box1 = alt.Chart(new_df).mark_boxplot().encode(x=alt.X(alt.repeat("column"),type='quantitative'),y=alt.Y("DEATH_EVENT",type='nominal')).repeat(column=quantitative_columns[:2])
box2 = alt.Chart(new_df).mark_boxplot().encode(x=alt.X(alt.repeat("column"),type='quantitative'),y=alt.Y("DEATH_EVENT",type='nominal')).repeat(column=quantitative_columns[2:4])
box3 = alt.Chart(new_df).mark_boxplot().encode(x=alt.X(alt.repeat("column"),type='quantitative'),y=alt.Y("DEATH_EVENT",type='nominal')).repeat(column=quantitative_columns[4:6])
box4 = alt.Chart(new_df).mark_boxplot().encode(x=alt.X(alt.repeat("column"),type='quantitative'),y=alt.Y("DEATH_EVENT",type='nominal')).repeat(column=quantitative_columns[6:])

box1 & box2 & box3 & box4
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

### Standardization

```python
scaler = StandardScaler()
x_train_scaled  = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_s = x_train[["ejection_fraction","serum_sodium","time"]]
x_test_s = x_test[["ejection_fraction","serum_sodium","time"]]
x_train_scaled_s  = scaler.fit_transform(x_train_s)
x_test_scaled_s = scaler.transform(x_test_s)
```

## Feature selection

### Z-score

```python
df_columns = list(data.columns)
z = {}
for i in df_columns:
    u0,sigma0 = norm.fit(data.loc[y==0,i])
    u1,sigma1 = norm.fit(data.loc[y==1,i])
    z[i] = np.abs((u0-u1))/np.sqrt(np.abs(sigma0**2/len(data.loc[y==0])-sigma1**2/len(data.loc[y==1])))
Z = pd.Series(z)
Z.sort_values(ascending=False)
```

### Boosting

```python
model_XGB=XGBClassifier(colsample_bytree=0.6, learning_rate=0.1, max_depth= 3, n_estimators=150, subsample= 0.8)
model_XGB.fit(x_train_scaled,y_train)
y_pred=model_XGB.predict_proba(x_test_scaled)
cm =confusion_matrix(y_test,y_pred[:,1]>0.5)
cmp = ConfusionMatrixDisplay(cm)
cmp.plot()
print("The accurancy of test data is {}".format(accuracy_score(y_test,y_pred[:,1]>0.5)))
print("The F1 Score of test data is {}".format(f1_score(y_test,y_pred[:,1]>0.5)))
print("The ROC AUC of test data is {}".format(roc_auc_score(y_test,y_pred[:,1])))
print("The Log Loss of test data is {}".format(log_loss(y_test,y_pred[:,1])))
```

```python
feature_importances = pd.Series(model_XGB.feature_importances_, index=x_train.columns).sort_values(ascending=False)
feature_importances.plot.bar()
```

```python
param_grid = {
    'n_estimators': [50,100, 150],  # Number of boosting rounds
    'max_depth': [3, 4],  # Maximum tree depth
    'learning_rate': [0.01, 0.1],  # Learning rate
    'subsample': [0.6, 0.8],  # Subsample ratio of the training instances
    'colsample_bytree': [0.6, 0.8],  # Subsample ratio of columns when constructing each tree
}


# Using GridSearchCV for parameter search
Gcv = GridSearchCV(estimator=model_XGB, param_grid=param_grid, cv=5, scoring='accuracy')

# Fitting the data
Gcv.fit(x_train,y_train)
# Outputting the best parameters
print("Best parameters found: ", Gcv.best_params_)
```

## GAMS

```python
df_list = [3,3 , 4, 4, 4, 3, 4]
degree_list = [2, 2, 3, 3, 3, 2, 3]
```

```python
x_train_gams = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_gams = pd.DataFrame(x_test_scaled,columns=x_train.columns)
```

```python
import statsmodels.gam.api as smg
import statsmodels.api as sm
bs = smg.BSplines(x_train_gams[quantitative_columns], df=df_list, degree=degree_list)
category = pd.get_dummies(x_train_gams[qualitative_columns],drop_first=True)
X = sm.add_constant(category)
gam_bs = smg.GLMGam(y_train, X, smoother=bs,family=sm.families.Binomial()) 
res_bs = gam_bs.fit()
res_bs.summary()
```

```python
quantitative_columns
```

```python
train_min = np.min(x_train_gams)
train_max = np.max(x_train_gams)
x_test_clipped = np.clip(x_test_gams, train_min, train_max,axis=1)

category = pd.get_dummies(x_test_clipped[qualitative_columns],drop_first=True)
X_test = sm.add_constant(category)

y_pred = res_bs.predict(X_test,x_test_clipped[quantitative_columns])
```

```python
cm =confusion_matrix(y_test,y_pred>0.5)
cmp = ConfusionMatrixDisplay(cm)
cmp.plot()
print("The accurancy of test data is {}".format(accuracy_score(y_test,y_pred>0.5)))
print("The F1 Score of test data is {}".format(f1_score(y_test,y_pred>0.5)))
print("The ROC AUC of test data is {}".format(roc_auc_score(y_test,y_pred)))
print("The Log Loss of test data is {}".format(log_loss(y_test,y_pred)))
```
